```python
def male_female_child(passenger):
    age, sex = passenger
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
tit_df['person'] = tit_df[['Age', 'Sex']].apply(male_female_child, axis=1)
```

```python
tit_df.head()
```

```python
sns.factorplot('Pclass', data=tit_df, hue='person', kind='count')
```

```python
# What deck were the passengers on and how does that relate to their class?
deck = tit_df['Cabin'].dropna()
```

```python
deck.head()
```

```python
levels = [level[0] for level in deck]
```

```python
levels[0:5]
```

```python
cabin_df = pd.DataFrame(levels)
```

```python
cabin_df.columns = ['Cabin']
```

```python
sns.factorplot('Cabin', data=cabin_df, palette='winter_d', kind='count', order=list('ABCDEFGT'))
```

```python
cabin_df = cabin_df[cabin_df.Cabin != 'T']
```

```python
sns.factorplot('Cabin', data=cabin_df, palette='winter_d', kind='count', order=list('ABCDEFG'))
```

```python
# Where did the passengers come from?
sns.factorplot('Embarked', data=tit_df, hue='Pclass', kind='count', order=list('CQS'))
```

```python
tit_df.head()
```

```python
# Who was alone and who was with family?
```

```python
tit_df['Alone'] = tit_df.SibSp + tit_df.Parch
tit_df['Alone'].loc[tit_df['Alone'] > 0] = 'With Family'
tit_df['Alone'].loc[tit_df['Alone'] == 0] = 'Alone'
```

```python
tit_df.head()
```

```python
sns.factorplot('Alone', data=tit_df, kind='count')
```

```python
# What factors helped someone survive the sinking?
```

```python
tit_df['Survivor'] = tit_df.Survived.map({0:'no', 1:'yes'})
sns.factorplot('Survivor', data=tit_df, kind='count', palette='Set1')
```

```python
sns.factorplot('Pclass', data=tit_df, hue='Survivor', kind='count')
```

```python
sns.factorplot('Pclass', 'Survived', hue='person', data=tit_df)
```

```python
sns.lmplot('Age', 'Survived', data=tit_df)
```

```python
generations = [10,20,40,60,80]
sns.lmplot('Age', 'Survived', hue='Pclass', data=tit_df, x_bins=generations)
```
