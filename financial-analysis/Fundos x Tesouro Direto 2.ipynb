```python
resp = requests.get('http://www.xpi.com.br/investimentos/fundos-de-investimento/detalhes-de-fundos-de-investimento.aspx?F=3409')
```

```python
start = 2017
series = []
for df in pd.read_html(resp.text, match='Janeiro', index_col=0, decimal=',', thousands='.'):
    fundo = df.ix['Fundo']
    fundo.name = '%s' % start
    start -= 1
    series.append(fundo)
```

```python
XP_FI = DataFrame(series)
```

```python
resp = requests.get('http://idg.receita.fazenda.gov.br/orientacao/tributaria/pagamentos-e-parcelamentos/taxa-de-juros-selic')
```

```python
dfs = []
for df in pd.read_html(resp.text, match='Janeiro', index_col=0, header=0, decimal=',', thousands='.'):
    dfs.append(df)

Selic = pd.concat(dfs, axis=1)
Selic
```

```python
Selic.drop([str(y) for y in range(1995, 2009)], axis=1, inplace=True)
Selic
```

```python
Selic = Selic.T
Selic
```

```python
Selic = Selic.replace('%', '', regex=True).replace(',', '.', regex=True).astype('float')
```

```python
Selic['Media'] = Selic.iloc[:, range(12)].mean(axis=1)
Selic['Acumulado'] = Selic.iloc[:, range(12)].sum(axis=1)
Selic
```

```python
to_plot = DataFrame(columns=['Selic', 'XP_FI'], index=Selic.index)
to_plot['Selic'] = Selic['Acumulado']
to_plot['XP_FI'] = XP_FI['Acumulado']
to_plot
```

```python
to_plot.plot(figsize=(11,5), legend=True)
```
